```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse = pd.read_csv(mouse_drug_data_to_load)
clinical = pd.read_csv(clinical_trial_data_to_load)
#clinical.head()

# Combine the data into a single dataset
pharma = pd.merge(clinical, mouse, on="Mouse ID", how="left")

# Display the data table for preview
pharma.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
group_pharma = pharma.groupby(["Drug", "Timepoint"])

# Convert to DataFrame
tumor_vol = group_pharma["Tumor Volume (mm3)"].mean()
tumor_volume_df = pd.DataFrame({"Tumor Volume (mm3)": tumor_vol})

# Preview DataFrame
tumor_volume_df.reset_index(inplace=True)
tumor_volume_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_vol_sem = group_pharma["Tumor Volume (mm3)"].sem()

# Convert to DataFrame
tumor_volume_sem_df = pd.DataFrame({"Tumor Volume (mm3)": tumor_vol_sem})

# Preview DataFrame
tumor_volume_sem_df.reset_index(inplace=True)
tumor_volume_sem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_volume_trans_df = tumor_volume_df.groupby(["Timepoint", "Drug"]).mean().unstack()
tumor_volume_sem_trans_df = tumor_volume_sem_df.groupby(["Timepoint", "Drug"]).mean().unstack()

# Preview that Reformatting worked
tumor_volume_trans_df.head()
```

```python
# Generate the Plot (with Error Bars)
timep = tumor_volume_trans_df.reset_index()["Timepoint"]

fig, ax = plt.subplots()

ax.errorbar(timep, tumor_volume_trans_df["Tumor Volume (mm3)"]["Capomulin"], 
            yerr = tumor_volume_sem_trans_df["Tumor Volume (mm3)"]["Capomulin"],
            color='red', marker='o', label='Capomulin', linewidth=1, linestyle='dotted')

ax.errorbar(timep, tumor_volume_trans_df["Tumor Volume (mm3)"]["Infubinol"], 
            yerr = tumor_volume_sem_trans_df["Tumor Volume (mm3)"]["Infubinol"],
            color='blue', marker='^', label='Infubinol', linewidth=1, linestyle='dotted')

ax.errorbar(timep, tumor_volume_trans_df["Tumor Volume (mm3)"]["Ketapril"], 
            yerr = tumor_volume_sem_trans_df["Tumor Volume (mm3)"]["Ketapril"], 
            color='green', marker='s', label='Ketapril', linewidth=1, linestyle='dotted')

ax.errorbar(timep, tumor_volume_trans_df["Tumor Volume (mm3)"]["Placebo"], 
            yerr = tumor_volume_sem_trans_df["Tumor Volume (mm3)"]["Placebo"], 
            color='black', marker='d', label='Placebo', linewidth=1, linestyle='dotted')

ax.set_title("Tumor Response to Treatment")
ax.set_xlabel("Time (Days)")
ax.set_ylabel("Tumor Volume ($mm^3$)")
ax.grid()
ax.legend()

#plt.tight_layout()

# Save the Figure
plt.savefig("../Images/Tumor_Volume.png")

# Show the Figure
plt.show()
```

## Analisys of tumor growth (tumor volume ($mm^3$))
* Emphasis in four main drugs (Capomulin, Infubinol, Ketapril and Placebo).
* From this group of drugs, only Capomulin allowed the tumor to decrease (Ramicane also worked, although is not in the group).
* The tumor response to the other drugs presented a similar performance.
* The standard errors shows that means (tumor volume) for each timepoint are consistent, and further analisys can be made.

```python
# Generate the Plot (with Error Bars)
timep = metast_sites_trans_df.reset_index()["Timepoint"]

fig, ax = plt.subplots()

ax.errorbar(timep, metast_sites_trans_df["Metastatic Sites"]["Capomulin"], 
            yerr = metast_sites_sem_trans_df["Metastatic Sites"]["Capomulin"],
            color='red', marker='o', label='Capomulin', linewidth=1, linestyle='dotted')

ax.errorbar(timep, metast_sites_trans_df["Metastatic Sites"]["Infubinol"], 
            yerr = metast_sites_sem_trans_df["Metastatic Sites"]["Infubinol"],
            color='blue', marker='^', label='Infubinol', linewidth=1, linestyle='dotted')

ax.errorbar(timep, metast_sites_trans_df["Metastatic Sites"]["Ketapril"], 
            yerr = metast_sites_sem_trans_df["Metastatic Sites"]["Ketapril"], 
            color='green', marker='s', label='Ketapril', linewidth=1, linestyle='dotted')

ax.errorbar(timep, metast_sites_trans_df["Metastatic Sites"]["Placebo"], 
            yerr = metast_sites_sem_trans_df["Metastatic Sites"]["Placebo"], 
            color='black', marker='d', label='Placebo', linewidth=1, linestyle='dotted')

ax.set_title("Metastatic Spread During Treatment")
ax.set_xlabel("Treatment Duration (Days)")
ax.set_ylabel("Met. Sites")
ax.grid()
ax.legend()

#plt.tight_layout()

# Save the Figure
plt.savefig("../Images/Metastatic_Sites.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice = group_pharma["Mouse ID"].count()

# Convert to DataFrame
mice_df = pd.DataFrame({"Mouse Count": mice})

# Preview DataFrame
mice_df.reset_index(inplace=True)
mice_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mice_trans_df = mice_df.groupby(["Timepoint", "Drug"]).mean().unstack()

# Preview the Data Frame
mice_trans_df.head()
```

```python
# Generate the Plot (Accounting for percentages)
timep = mice_trans_df.reset_index()["Timepoint"]

fig, ax = plt.subplots()

ax.errorbar(timep, (mice_trans_df["Mouse Count"]["Capomulin"] / mice_trans_df["Mouse Count"]["Capomulin"][0]) * 100, 
            color='red', marker='o', label='Capomulin', linewidth=1, linestyle='dotted')

ax.errorbar(timep, (mice_trans_df["Mouse Count"]["Infubinol"] / mice_trans_df["Mouse Count"]["Infubinol"][0]) * 100, 
            color='blue', marker='^', label='Infubinol', linewidth=1, linestyle='dotted')

ax.errorbar(timep, (mice_trans_df["Mouse Count"]["Ketapril"] / mice_trans_df["Mouse Count"]["Ketapril"][0]) * 100,  
            color='green', marker='s', label='Ketapril', linewidth=1, linestyle='dotted')

ax.errorbar(timep, (mice_trans_df["Mouse Count"]["Placebo"] / mice_trans_df["Mouse Count"]["Placebo"][0]) * 100,  
            color='black', marker='d', label='Placebo', linewidth=1, linestyle='dotted')

ax.set_title("Survival During Treatment")
ax.set_xlabel("Time (Days)")
ax.set_ylabel("Survival Rate (%)")
ax.grid()
ax.legend()

# Save the Figure
plt.savefig("../Images/Survival_Rate.png")

# Show the Figure
plt.show()
```

## Analisys of survival rate
* For each drugs were selected 25 mice. After 45 days of treatment those whose drug applied was Capomulin and Ramicane survived in 84% approximately. For the rest of mice the survival rate after 45 days was in a range 36-56%.
* Definitely the aplicaction of any of those does not guarantee the survival of all the mice. Maybe it depends also on others factors.

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
group_pharma = pharma.groupby(["Drug", "Timepoint"])
tumor_change = pd.DataFrame()
tumor_change["Tumor Volume"] = group_pharma['Tumor Volume (mm3)'].mean()
tumor_change = tumor_change.unstack(level=-1)
tumor_change.columns = tumor_change.columns.droplevel(level=0)
tumor_change = (tumor_change[45] - tumor_change[0]) / tumor_change[0] * 100

# Display the data to confirm
tumor_change
```

```python
# Store all Relevant Percent Changes into a Tuple
relevant_drugs = ("Capomulin", "Infubinol", "Ketapril", "Placebo")

changes_relevant = tuple(tumor_change[relevant] for relevant in relevant_drugs)

# Splice the data between passing and failing drugs
colors = []
for i in changes_relevant:
    if i < 0:
        colors.append("green")
    else:
        colors.append("red")

# Orient widths. Add labels, tick marks, etc. 
fig, ax = plt.subplots()

rel_dru = np.arange(len(relevant_drugs))
width = 0.99
changes = ax.bar(rel_dru, changes_relevant, color=colors, width=width)

ax.set_xlim((-0.5, 3.5))
ax.set_ylim((-25, 60))
ax.set_ylabel('% Tumor Volume Change')
ax.set_title('Tumor Change Over 45 Day Treament')
ax.set_xticks(rel_dru + width/2)
ax.set_yticks(range(-20, 70, 20))
ax.set_xticklabels(relevant_drugs)
ax.grid()


# Use functions to label the percentages of changes
def autolabel(change, xpos='center'):
    posx = range(len(relevant_drugs))
    posy = []
    for i in changes_relevant:
        if i >= 0:
            posy.append(3)
        else:
            posy.append(-4)
            
    va = []
    for i in changes_relevant:
        if i >= 0:
            va.append("bottom")
        else:
            va.append("top")
        
    labels = changes_relevant
    for i in range(len(relevant_drugs)):
        ax.text(posx[i], posy[i], '%1.0d%%' % int(labels[i]), ha='center', va=va[i], color='white')

# Call functions to implement the function calls
autolabel(changes)


# Save the Figure
plt.savefig("../Images/per_change.png")

# Show the Figure
fig.show()
```

## Analisys of summary bar chart
* For the four relevant drugs the tumor volumen decreased 19% with the application of Capomulin and increased in about 55% for the other drugs.
